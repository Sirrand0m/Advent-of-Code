```python
with open("Day 5 Input.txt") as f:
    database = f.readlines()
database = [line.strip() for line in database]
fresh_ranges = database[:190]
ingredient_ids = [int(i) for i in database[191:]]
test_ranges = ["3-5",
"10-14",
"16-20",
"12-18"]
test_ids = [1,
5,
8,
11,
17,
32]
```

```python
def part_2(ranges):
    fresh_ids = []
    for freshness_range in ranges:
        freshness_range_split = freshness_range.split("-")
        starting = int(freshness_range_split[0])
        ending = int(freshness_range_split[1])+1
        fresh_ids.append(range(starting, ending))

    fresh_ids.sort(key=lambda x:x[0])
    flag = True
    while flag:
        for i in range(len(fresh_ids) -1):
            flag = False
            r1, r2 = fresh_ids[i].start, fresh_ids[i].stop
            r3, r4 = fresh_ids[i+1].start, fresh_ids[i+1].stop
            if r2 < r3:
                continue
            fresh_ids[i] = range(min(r1,r3),max(r2,r4))
            fresh_ids.pop(i+1)
            flag = True
            break

    lengths = [len(i) for i in fresh_ids]

    return sum(lengths)
```
