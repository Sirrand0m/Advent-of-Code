```python
test_warehouse = ["..@@.@@@@.",
"@@@.@.@.@@",
"@@@@@.@.@@",
"@.@@@@..@.",
"@@.@@@@.@@",
".@@@@@@@.@",
".@.@.@.@@@",
"@.@@@.@@@@",
".@@@@@@@@.",
"@.@.@@@.@."]
```

```python
test_locs = set()
for i, row in enumerate(test_warehouse):
    for j, element in enumerate(row):
        if element == "@":
            test_locs.add((i,j))
```

```python
test_locs.__len__()
```

```python
def count_neighbours(loc_set):
    count_dict = {}
    from itertools import product
    for loc in loc_set:
        count = 0
        x,y = loc[0], loc[1]
        x_coords, y_coords = [x-1, x, x+1], [y-1, y, y+1]
        neighbours = product(x_coords, y_coords)
        for neighbour in neighbours:
            if neighbour in loc_set and neighbour != (x,y):
                count += 1
        count_dict[loc] = count
    return count_dict
```

```python
test_dict = count_neighbours(test_locs)
```

```python
movable_count = 0
for coord in test_dict.values():
    if coord < 4:
        movable_count += 1
movable_count
```

```python
part_1_dict = count_neighbours(roll_locs)
```

```python
movable_count = 0
for coord in part_1_dict.values():
    if coord < 4:
        movable_count += 1
movable_count
```

```python
roll_locs_p2 = roll_locs.copy()
```

```python
def part_2(loc_dict, loc_set):
    loc_set_copy = loc_set.copy()
    total_movable_count = 0
    prev_iter_movable_count = -9999
    while prev_iter_movable_count != total_movable_count:
        movable_count = 0
        for coord, neighbours in loc_dict.items():
            if neighbours < 4:
                movable_count += 1
                loc_set_copy.remove(coord)
        prev_iter_movable_count = total_movable_count
        total_movable_count += movable_count
        loc_dict = count_neighbours(loc_set_copy)
    return total_movable_count
```

```python
part_2(test_dict, test_locs)
```

```python
part_2(part_1_dict, roll_locs_p2)
```
