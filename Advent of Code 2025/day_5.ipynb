```python
test_ranges = ["3-5",
"10-14",
"16-20",
"12-18"]
test_ids = [1,
5,
8,
11,
17,
32]
```

```python
def check_freshness(ranges, id):
    """Given ranges, check if a single ID is in any of the ranges"""
    for freshness_range in ranges:
        freshness_range_split = freshness_range.split("-")
        starting = int(freshness_range_split[0])
        ending = int(freshness_range_split[1])+1
        if id in range(starting, ending):
            return 1
    return 0
```

```python
def part_1(ranges, ids):
    fresh_list = [check_freshness(ranges, i) for i in ids]
    return sum(fresh_list)
```

```python
part_1(test_ranges, test_ids)
```

```python
part_1(fresh_ranges, ingredient_ids)
```

```python
from itertools import combinations
```

```python
def part_2(ranges):
    fresh_ids = []
    for freshness_range in ranges:
        freshness_range_split = freshness_range.split("-")
        starting = int(freshness_range_split[0])
        ending = int(freshness_range_split[1])+1
        fresh_ids.append(range(starting, ending))
    fresh_ids, flag = check_combis(fresh_ids)

    while flag:
        fresh_ids, flag = check_combis(fresh_ids)
        print(len(fresh_ids))

    lengths = [len(i) for i in fresh_ids]

    return sum(lengths)

def range_intersect(r1, r2):
    return range(max(r1.start,r2.start), min(r1.stop,r2.stop)) or None

def check_combis(fresh_ids):
    flag = False
    for i,j in combinations(fresh_ids, 2):
        if range_intersect(i,j):
            flag = True
            fresh_ids.append(range(min(i.start,j.start), max(i.stop,j.stop)))
            try:
                fresh_ids.remove(i)
            except ValueError:
                pass
            try:
                fresh_ids.remove(j)
            except ValueError:
                pass
    return fresh_ids, flag
```

```python
part_2(test_ranges)
```

```python
part_2(fresh_ranges)
```
