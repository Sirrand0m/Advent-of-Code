```python
ids = ids[0].split(",")
```

```python
test_ids = ["11-22","95-115","998-1012","1188511880-1188511890","222220-222224",
"1698522-1698528","446443-446449","38593856-38593862","565653-565659",
"824824821-824824827","2121212118-2121212124"]
```

```python
def part_1(ids):
    sum_ids = 0

    for id_range in ids:
        id_range = id_range.split("-")
        start_id = int(id_range[0])
        end_id = int(id_range[1])
        
        # check for invalid ids
        for prod_id in range(start_id, end_id+1):
            if invalid_id_checker(prod_id):
                sum_ids += prod_id
    return sum_ids

def invalid_id_checker(prod_id):
    prod_id = str(prod_id)
    length = len(prod_id)
    if length % 2 == 1:
        return False
    if prod_id[:length//2] == prod_id[length//2:]:
        return True
    else:
        return False
```

```python
part_1(test_ids)
```

```python
part_1(ids)
```

```python
def part_2(ids):
    sum_ids = 0

    for id_range in ids:
        id_range = id_range.split("-")
        start_id = int(id_range[0])
        end_id = int(id_range[1])
        
        # check for invalid ids
        for prod_id in range(start_id, end_id+1):
            if invalid_id_checker(prod_id):
                sum_ids += prod_id
                continue
            if new_invalid_id_checker(prod_id):
                sum_ids += prod_id
    return sum_ids

def new_invalid_id_checker(prod_id):
    prod_id = str(prod_id)
    length = len(prod_id)

    # edge case: single digit
    if length == 1:
        return False

    # check length 1
    if prod_id[0]*length == prod_id:
        return True

    # check lengths up to half the id length
    for l in range(2,length//2 + 1):
        if length % l:
            continue
        seq = prod_id[:l]
        multiple = length//l
        if seq*multiple == prod_id:
            return True
    return False
```

```python
part_2(test_ids)
```

```python
part_2(ids)
```
