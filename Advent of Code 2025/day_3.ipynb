```python
with open("Day 3 Input.txt") as f:
    banks = f.readlines()

len(banks)
```

```python
test_bank = ["987654321111111", "811111111111119", "234234234234278", "818181911112111"]
```

```python
banks = [bank.strip() for bank in banks]
```

```python
def part_1(bank_list):
    output_joltage = 0
    for bank in bank_list:
        max_ten = "0"
        max_one = "0"
        for ind, tens_digit in enumerate(bank[:-1]):
            if int(tens_digit) > int(max_ten):
                max_ten = tens_digit
                max_one = "0"
                for ones_digit in bank[ind+1:]:
                    if int(ones_digit) > int(max_one):
                        max_one = ones_digit
        output_joltage += int(max_ten + max_one)
    return output_joltage
        
```

```python
part_1(banks)
```

```python
def part_2(bank_list):
    output_joltage = 0
    for bank in bank_list:
        output_joltage += find_max_joltage(bank)
    return output_joltage

def find_max_joltage(bank):
    batteries = [int(b) for b in bank]
    digits = []
    start = 0
    joltage = ""

    for d in range(12):
        end = -(12 - d - 1)
        if end == 0:
            end = None
        sub = batteries[start:end]
        max_val = max(sub)
        start = start + sub.index(max_val) + 1
        digits.append(max_val)

    for d in digits:
        joltage += str(d)

    return int(joltage)
```

```python
part_2(test_bank)
```
